```
df_=df.copy()#future use
#Go through each row of the dataframe and delete it if any of the columns is null
#method-1
for r in df.index:
  for c in df.columns:
    if pd.isnull(df.loc[r,c]):
      df.drop(r,inplace=True)
      break;
df.describe()
```

```
#filtering the data
percentile=df.distance.quantile(0.75)
dt=df.loc[(df["year"]>=2010)&
          ((dt["method"]=="Transit") | (dt["method"]=="Radial Velocity")) & 
          (dt["distance"]>percentile)]
dt.describe()
```

```
df.replace(to_replace="Radial Velocity",value="rv",inplace=True)
df
```

```
df.loc[:,"method"].unique()
```

```
short_names={}
for s in df.loc[:,"method"].unique():
  short_names[s]=''.join([x[0] for x in s.split(" ")])
print(short_names)
```

```
for i,r in df.iterrows():
  df.loc[i,"short_method"]=short_names.get(r["method"],r["method"])
```

```
df.drop("method",axis=1)
```

```
def shorten(s):
  return short_names.get(s,s)
df=sns.load_dataset('planets')#reseting
df.loc[:,"short"]=df.loc[:,"method"].apply(shorten)
df
```
